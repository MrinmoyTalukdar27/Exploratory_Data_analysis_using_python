```python
plt.figure(figsize =(4,4))
ax = sns.countplot(x= "SeniorCitizen", data =df,color="yellow")
ax.bar_label(ax.containers[0])
plt.title("Churn By seniorcitizen")
plt .show()
```

# In the above
* we are checking how many coustomers are Senior Citizen

```python
# Group by SeniorCitizen and Churn
senior_churn = df.groupby(["SeniorCitizen", "Churn"]).size().unstack(fill_value=0)

# Convert to percentages per SeniorCitizen group
senior_churn_pct = senior_churn.div(senior_churn.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = senior_churn_pct.plot(kind='bar', stacked=True, figsize=(5,5), color=["yellow", "red"])

# Add percentage labels inside bars
for c in ax.containers:
    labels = [f'{v.get_height():.1f}%' if v.get_height() > 0 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='center', fontsize=10)

# Titles and labels
plt.title("Churn by Senior Citizen (Stacked Bar Chart)", fontsize=10)
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn Status", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# From the Above Analysis We  Can See 
* Coparatively a Large Number of Pepol in the Senior_citizen Catagory have Churned out

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]
fig, axes = plt.subplots(nrows=(len(cols) + 2)//3, ncols=3, figsize=(18, 12))  # 3 plots per row

axes = axes.flatten()  # Flatten 2D axes array for easy looping

# Loop through each column and axis
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, hue="Churn", palette={"Yes": "red", "No": "yellow"}, ax=axes[i])
    axes[i].set_title(f"{col} vs Churn")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=45)  # ✅ rotate labels safely

# Hide any unused subplot boxes (if total plots < grid size)
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

# From The Above Analysis We Can Say That
* Customers who do not use additional services like online security, backup, or tech support tend to have a higher 
churn rate.
* Fiber optic users show more churn compared to DSL users, while customers with extra services are more likely to stay.
* Overall, having more support and protection services reduces the chances of churn.
